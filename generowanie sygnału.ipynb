Przebieg sinusoidalny
Innym podstawowym sygnałem jest fala sinusoidalna. Trzy parametry opisują rzeczywisty sygnał sinusoidalny:
Amplituda (A), częstotliwość (pulsacja) (ω0) oraz faza (φ).

# Przebieg sinusoidalny - sygnał ciągły

```python
from math import pi, sin, cos

def sinus(A, tp, tk, fp, f, phi):
    x = np.arange(tp, tk, 1/fp)
    y = [A*sin(2*pi*f*t) for t in x]
    y = np.array(y)
    return x, y



x1, y1 = sinus(1, 0, 1, 256, 10, 0)

x2, y2 = sinus(1, 0, 1, 256, 15, pi/4)

y = y1+y2
Y = [f + np.random.randn(1) for f in y]

fig, axs = plt.subplots(4)
fig.suptitle('Signals')
axs[0].plot(x1, y1)
axs[0].set_title('Sygnał sinusoidalny')
axs[1].plot(x1, y2)
axs[1].set_title('Sygnał sinusoidalny')
axs[2].plot(x1, y)
axs[2].set_title('Suma sygnałów')
axs[3].plot(x1, Y)
axs[3].set_title('Suma sygnałów z szumem gaussowskim')

```

b) Za pomocą napisanej funkcji wygeneruj sygnał, który:

trwa 1 s,
jego częstotliwość wynosi 10 Hz,
jego częstotliwość próbkowania wynosi 256 Hz,
jego amplituda wynosi 2,
przesunięcie sygnału to pi/4. 


```python
def cosinus(A, f, tp, tk, fp, phi):
    """
    Ta funkcja zwraca dwa wektory - wektor czasu oraz wektor sygnału 
    
    Dane wejściowe:
    A - amplituda sygnału
    f - częstotliwość sygnału
    tp - czas początkowy sygnału
    tk - czas końcowy sygnału
    fp - częstotliwość próbkowania sygnału
    phi - przesunięcie sygnału
    
    Dane wyjściowe:
    x - wektor czasu
    y - sygnał
    """
    
    x = np.arange(tp, tk, 1/fp)
    y = [cos(2*pi*f*T+phi) for T in x]
    return x, y

x, y = cosinus(2, 10, 0, 1, 256, pi/4)

plt.plot(x, y)

for i in range(len(x)):
    if y[i] == 1:
        print(i)
        
print(x[48])
print(x[76])
print(pi/12)
print(len(x))
```
